```python
df1 = df.drop_duplicates(subset='e', keep='first')
df1 = df1.drop_duplicates(subset='g', keep='first')
df1
```

```python
def remove_non_decreasing(df):
    
    i=0
    while i+1<len(df):
        if df.iloc[i]['g'] <= df.iloc[i+1]['g']:
            df = df.drop(df.index[i])
        i = i+1    
    return df

df1 = remove_non_decreasing(df1)
len(df1)
```
